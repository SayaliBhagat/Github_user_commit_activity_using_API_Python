```python
import json, requests 
from pprint import pprint
import base64
from github import Github
from pprint import pprint
url = f"https://api.github.com/repos/vuejs/vue/stats/contributors"
user_data = requests.get(url).json()
pprint(user_data)
```

```python
all_contributors = list()
page_count = 1
while True:
    contributors = requests.get("https://api.github.com/repos/vuejs/vue/stats/contributors?page=%d"%page_count)
    if contributors != None and contributors.status_code == 200 and len(contributors.json()) > 0:
        all_contributors = all_contributors + contributors.json()
    else:
        break
    page_count = page_count + 1
count=len(all_contributors)
print("Total number of active contributors %d" %count)
```
